#### Scrivere un programma ch presenta un menu di scelta come di seguito riportato

1. Somma tra 2 numeri
2. Concatenare 2 stringhe date in input
3. Pulire la consolle di debug
4. Terminare il programma

#### Il menu di scelta deve essere presentato a video fin quando l'utente non decide di terminare il programma
<hr>

## Funzioni per operazione

```python
import os

# Somma
def somma ():
    print('\n--- SOMMA TRA 2 NUMERI ---')
    try:
        num1 = float(input('Inserisci il primo numero: '))
        num2 = float(input('Inserisci il secondo numero: '))
        print('La somma:', num1 + num2)
    except ValueError:
        print('⚠️ Errore: devi inserire solo numeri')

# Concatenare 2 stringhe date in input
def concatenazione():
    print('\n--- COMCATENAZIONE DI STRINGHE ---')
    try:
        stringa1 = input('Inserisci la prima stringa: ')
        stringa2 = input('Inserisci la seconda stringa: ')

        # Verifica: Non devono essere numeri
        if stringa1.isdigit() or stringa2.isdigit():
            print('Errore: non puoi inserire numeri! Devi inserire solo stringhe🤨')
            return

        print('Risultato: ', stringa1 + ' ' + stringa2)
        # print(f'Risultato: {stringa1} - {stringa2}')
        # print('Risultato:', stringsa1, stringa2)
    except Exception as ex:
        print('⚠️ Errore inattesa', ex)


# Pulire la consolle di debug
def pulisci_console():
    os.system('cls' if os.name == 'nt' else 'clear')


def mostra_menu():
    print('\n=== MENU DI SCELTA ===')
    print("1. SOMMA TRA 2 NUIMERI")
    print("2. CONCARENARE 2 STRINGHE")
    print("3. PULISCI LA CONSOLLE SI DEBUG")
    print("4. ESCI DAL PROGRAMMA")
```

<h1><i>Programma principale</i></h1>

```python
def main():
    while True:
        mostra_menu()

        # variabile di scelt
        scelta = int(input('Scegli un opzione: '))

        try:
            if scelta == 1:
                somma()
            elif scelta == 2:
                concatenazione()
            elif scelta == 3:
                pulisci_console()
            elif scelta == 4:
                print('\nProgramma treminato. Riprova quando vuoi')
                break
        except Exception as e:
            print('⚠️ Errore generale: ', e)
```

```python
main()
```
