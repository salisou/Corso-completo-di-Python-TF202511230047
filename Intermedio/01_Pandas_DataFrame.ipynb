```python
import pandas as pd

# Creazione della lista degli studenti con i voti
studenti = {
    'nome': ['Massa', 'Anntonella', 'Luigi', 'Eduardo', 'Matteo', 'Stefano', 'Lucia', 'Annaritantonia', 'Paolo', 'Francesca'],
    'voti':  [10, 8, 9, 7, 10, 6, 5, 4, 10, 8]
}


# Creazione del DataFrame(Foglio excel, Tabella, Entità)
df = pd.DataFrame(studenti)

print(df)
```

La funzione info() restituisce l'infromazione del DataFrame() 

```python
df.info()
```

le funzione di agreazione del DataFrame()
- count  => somma 
- mean   => media
- min    => minimo
- max    => massimo

```python
print(df.describe())
```

# Filtrare dati
- Studenti promossi

```python
promossi = df[df['voti'] >= 9]
promossi
```

<h3><i>Calcolare la media</i></h3> 

```python
media = df['voti'].mean()
print(media)
```
